```python
class NetworkProcessor:
    def __init__(self, file_path):
      self.file_path = file_path # Read file
      self.lines_list = [] #line ,a b,d c
      #
      self.edges = [] # Split \n, \t
      self.nodes = []
      #
      self.self_loops = []
      self.multi_edges = []
      #
      self.all_neighbors = []
      self.self_neighbors = []
      self.duplicated_neighbors = []

    def read_file(self):
      with open(self.file_path, 'r') as file:
          line = file.readline()
          while line:
            self.lines_list.append(line)
            line = file.readline()

    def get_edges(self):
      edge = [item.strip('\n') for item in self.lines_list]
      self.edges = [item.split('\t') for item in edge]

    def get_unique_nodes(self):
      all_nodes = [item for sublist in self.edges for item in sublist]
      unique_nodes = list(set(all_nodes))
      self.nodes = sorted(unique_nodes)

    def get_simple_or_not(self):
      for start_node,end_node in self.edges:
        if start_node == end_node:
          self.self_loops.append((start_node,end_node))
        else:
          self.multi_edges.append((start_node,end_node))
      if self.self_loops == []:
        print('Network is Simple !!! \n')
        print('multi-edges is ',self.multi_edges)
      else :
        print('Network is not Simple !!! \n')
        print('self-loops is ',self.self_loops)
        print('multi-edges is ',self.multi_edges)

    def get_neighbors(self, node):
      total_neighbor = []
      for start_node,end_node in self.edges:
        if (node == start_node) and (node == end_node): #all same
          self.self_neighbors.append(node)
          total_neighbor.append(node)
        elif (node != start_node) and (node == end_node): #end_node same
          total_neighbor.append(start_node)
        elif (node == start_node) and (node != end_node): #start_node same
          total_neighbor.append(end_node)
      for neighbor in total_neighbor: #unique or not
        if neighbor in self.all_neighbors:
          self.duplicated_neighbors.append(neighbor)
        else:
          self.all_neighbors.append(neighbor)

```

### 1. calculate the number of nodes and edges

```python
#network
network = NetworkProcessor('Network1.txt') #Network2.txt
network.read_file()
network.get_edges()
network.get_unique_nodes()
print("nodes =", len(network.nodes))
print("edges =", len(network.edges))
```
