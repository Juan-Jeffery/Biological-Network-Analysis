### Network answers

```python
import networkx as nx
import matplotlib.pyplot as plt
with open("/content/drive/MyDrive/生物網路/Network 1.txt") as f:
    lines = f.readlines()
network2_edge_list = []
for i in range(len(lines)):
    network2_edge_list.append(lines[i].strip().split("\t"))

print(network2_edge_list)
G = nx.Graph()
G.add_edges_from(network2_edge_list)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=12, font_color='black', font_weight='bold')
plt.show()
```
